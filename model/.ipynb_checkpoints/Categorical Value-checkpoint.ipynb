```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv('./../data/tic_tac_toe_dataset.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
categorical = ['top_left_square', 'top_middle_square', 'top_right_square','middle_left_square', 'middle_middle_square', 'middle_right_square','bottom_left_square', 'bottom_middle_square', 'bottom_right_square']
```

```python
X = data[categorical]
```

```python
X.head()
```

```python
X_new = pd.get_dummies(X)
```

```python
X_new.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb_make = LabelEncoder()
data["class"] = lb_make.fit_transform(data["class"])
```

```python
data.head()
```

```python
data.drop(categorical,inplace=True,axis=1)
```

```python
data.head()
```

```python
data.reset_index(drop = True, inplace= True)
X_new.reset_index(drop = True, inplace= True)
```
